```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv('/Users/Tobias/Documents/Studium/10SoSe_19_AI/Grundlagen_Sozialer_Netze_Schmidt/FrightDataAnalysis/Data/dot2_ytd_2007-2018_without2009_usa-mx.csv', low_memory=False)
```

```python
df7 = df[df["YEAR"] == 2007]
df8 = df[df["YEAR"] == 2008]
df10 = df[df["YEAR"]== 2010]
df11 = df[df["YEAR"] == 2011]
df12 = df[df["YEAR"] == 2012]
df13 = df[df["YEAR"] == 2013]
df14 = df[df["YEAR"] == 2014]
df15 = df[df["YEAR"] == 2015]
df16 = df[df["YEAR"] == 2016]
df17 = df[df["YEAR"] == 2017]
df18 = df[df["YEAR"] == 2018]

#Kontrolle
print(len(df7) + len(df8) + len(df10) + len(df11) + len(df12) + len(df13) + len(df14) + len(df15) + len(df16) + len(df17) + len(df18))
```

```python
def state_values(df_year, trdtype, year):
    states = ['AL', 'AK', 'AS', 'AZ', 'AR', 'CA', 'CO', 'CT', 'DE', 'DC', 'FL', 'GA', 'HI', 'ID', 'IL', 'IN', 'IA', 'KS', 'KY', 'LA', 'ME', 'MD', 'MA', 'MI', 'MN', 'MS', 'MO', 'MT', 'NE', 'NV', 'NH', 'NJ', 'NM', 'NY', 'NC', 'ND', 'OH', 'OK', 'OR', 'PA', 'RI', 'SC', 'SD', 'TN', 'TX', 'UT', 'VT', 'VA','WA', 'WV', 'WI', 'WY']
    x_pos = np.arange(len(states))
    performance = []

    i = 0

    while i<len(states):
        summe = 0
    
        for index, row in df_year.iterrows():  
                if row['USASTATE'] == states[i]:
                    summe += row['VALUE']
            
        performance.append(summe)
        i = i+1
    
    plt.figure(figsize=(40, 10))
    plt.bar(x_pos, performance, align='center', alpha=0.8)
    plt.xticks(x_pos, states)
    plt.ylabel('Value')
    plt.title(trdtype + ' per State ' + year)

    plt.show()
```
